```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_json(r'C:\Users\Ana\Desktop\Desafio_Engenharia_Dados\Atividade 03\NFe.json',orient = 'records')
df
```

```python
df_primeira_parte = df[['CreateDate','EmissionDate','Discount','NFeNumber','NFeID']]

df_NFeID = df['NFeID']

df_primeira_parte
```

```python
df_tratamento = df['ItemList']
df_tratamento
```

```python
df_tratamento[0]
```

```python
df_dict_final = pd.DataFrame()

for i in range(len(df_tratamento)):
    df_dict = df_tratamento[i]
    df_dict_transform = pd.DataFrame.from_dict(df_dict)
    df_dict_transform["NFeID"] = df_NFeID[i]
    df_dict_final = pd.concat([df_dict_final,df_dict_transform],ignore_index=True)

df_dict_final
```

```python
df_final = df_dict_final.merge(df_primeira_parte,on='NFeID')

df_ajuste_final = df_final[['CreateDate','EmissionDate','NFeNumber','NFeID','ProductName','Value','Quantity','Discount']]

df_ajuste_final
```
